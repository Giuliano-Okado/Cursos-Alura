```python
import pandas as pd
import numpy as np
import datetime
```

## PROJETO I

```python
dados_vendas_clientes = pd.read_json('dados_vendas_clientes.json')
dados_vendas_clientes.head()
```

```python
dados_vendas_clientes = pd.json_normalize(dados_vendas_clientes['dados_vendas'])
dados_vendas_clientes
```

```python
colunas = list(dados_vendas_clientes.columns)
colunas
```

```python
dados_vendas_clientes = dados_vendas_clientes.explode(colunas[1:])
dados_vendas_clientes
```

```python
dados_vendas_clientes.reset_index(inplace=True, drop=True)
dados_vendas_clientes
```

```python
dados_vendas_clientes.info()
```

```python
dados_vendas_clientes['Valor da compra'] = dados_vendas_clientes['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip()).astype(np.float64)
```

```python
dados_vendas_clientes.info()
```

```python
dados_vendas_clientes['Cliente'] = dados_vendas_clientes['Cliente'].str.lower()
dados_vendas_clientes['Cliente'] = dados_vendas_clientes['Cliente'].str.replace('[^a-z ]', '', regex=True).str.strip()
dados_vendas_clientes.head()
```

```python
dados_vendas_clientes['Data de venda'] = pd.to_datetime(dados_vendas_clientes['Data de venda'], format='%d/%m/%Y')
dados_vendas_clientes
```

```python
total_compras = dados_vendas_clientes.groupby(['Cliente'])['Valor da compra'].sum()
total_compras
```

## PROJETO II

```python
dados_locacao_imoveis = pd.read_json('dados_locacao_imoveis.json')
dados_locacao_imoveis.head()
```

```python
dados_locacao_imoveis = pd.json_normalize(dados_locacao_imoveis['dados_locacao'])
dados_locacao_imoveis
```

```python
dados_locacao_imoveis.info()
```

```python
colunas = list(dados_locacao_imoveis.columns)
colunas
```

```python
dados_locacao_imoveis = dados_locacao_imoveis.explode(colunas[1:])
dados_locacao_imoveis
```

```python
dados_locacao_imoveis.reset_index(inplace=True, drop=True)
dados_locacao_imoveis
```

```python
dados_locacao_imoveis.info()
```

```python
dados_locacao_imoveis['valor_aluguel'] = dados_locacao_imoveis['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',', '.').replace('reais', '').strip())
```

```python
dados_locacao_imoveis['valor_aluguel'] = dados_locacao_imoveis['valor_aluguel'].astype(np.float64)
```

```python
dados_locacao_imoveis.info()
```

```python
dados_locacao_imoveis['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao_imoveis['datas_combinadas_pagamento'], dayfirst=True)
dados_locacao_imoveis['datas_de_pagamento'] = pd.to_datetime(dados_locacao_imoveis['datas_de_pagamento'], dayfirst=True)
dados_locacao_imoveis['datas_combinadas_pagamento_formatted'] = dados_locacao_imoveis['datas_combinadas_pagamento'].dt.strftime('%m/%d/%Y')
```

```python
dados_locacao_imoveis.info()
```
